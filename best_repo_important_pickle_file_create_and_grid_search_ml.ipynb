```python
import numpy as np
import pickle
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier

# Load the dataset
df = pd.read_csv("/content/zoo.csv")

# Set target variable
y = df['domestic']

# Drop target variable and the "animal_name" column
df = df.drop(['domestic'], axis=1)

# One-hot encode the categorical features
df = pd.get_dummies(df)

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)

# Scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define the model
def create_model(optimizer='adam', units=16):
    model = Sequential()
    model.add(Dense(units, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(units, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Wrap the model in a KerasClassifier for use with grid search
model = KerasClassifier(build_fn=create_model)

# Define the grid search parameters
param_grid = {'units': [16, 32, 64, 128], 'optimizer': ['adam', 'rmsprop']}

# Perform grid search
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_search.fit(X_train, y_train)

# Print the best parameters
print("Best parameters: ", grid_search.best_params_)

# Save the model to a pickle file
with open('zoo_model.pkl', 'wb') as file:
    pickle.dump(grid_search, file)
filename='zoo_model.pkl'
# load the saved model and use it to perform inference on the test data
loaded_model = pickle.load(open(filename, 'rb'))
y_pred = loaded_model.predict(X_test)

# perform inference on the test data
print("Test set predictions:\n", y_pred)    
```
