```python
import cv2
import os

# Get list of all images in the folder
img_folder = "/content/images"
img_files = [f for f in os.listdir(img_folder) if f.endswith(".jpg")]
img_paths = [os.path.join(img_folder, f) for f in img_files]
images = [cv2.imread(img) for img in img_paths]

texts = ["Text 1", "Text 2", "Text 3", "Text 4", "Text 5"]

# Create a VideoWriter object
height, width, _ = images[0].shape
video = cv2.VideoWriter('vdfgdfgdfgideo.avi', cv2.VideoWriter_fourcc(*'XVID'), 20.0, (width, height))

# Add each image and its corresponding text to the video
for i in range(len(images)):
    image = images[i].copy()
    text = texts[i]
    # Get the size of the text
    text_size = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 1, 2)[0]
    # Put the text in the center of the image
    text_x = (width - text_size[0]) // 2
    text_y = (height + text_size[1]) // 2
    cv2.putText(image, text, (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    for j in range(20):
        video.write(image)

# Release the video writer
video.release()
```

```python
import cv2
import numpy as np

# Texts to be displayed
texts = ["Text 1", "Text 2", "Text 3", "Text 4", "Text 5"]

# Background image paths
bg_images = ["/content/images/bg1.jpg", "/content/images/bg2.jpg", "/content/images/bg3.jpg", "/content/images/bg4.jpg", "/content/images/bg5.jpg"]

# Video writer object to save the output video
video_writer = cv2.VideoWriter("/content/output.avi", cv2.VideoWriter_fourcc(*"MJPG"), 30, (1920, 1080))

# Loop through each text and background image
for i in range(len(texts)):
    # Read the background image
    bg_image = cv2.imread(bg_images[i])
    height, width, _ = bg_image.shape

    # Create a black image with the same size as the background image
    text_image = np.zeros((height, width, 3), dtype=np.uint8)

    # Add the text to the image
    text = texts[i]
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1.0
    thickness = 2
    text_size, baseline = cv2.getTextSize(text, font, font_scale, thickness)
    text_x = (width - text_size[0]) // 2
    text_y = (height + text_size[1]) // 2
    cv2.putText(text_image, text, (text_x, text_y), font, font_scale, (255, 255, 255), thickness)

    # Display the text over the background image for 10 seconds
    for j in range(300):
        frame = cv2.addWeighted(bg_image, 0.5, text_image, 0.5, 0)
        video_writer.write(frame)

    # Release the current frame
    cv2.destroyAllWindows()

# Release the video writer
video_writer.release()
```

```python
import cv2
import numpy as np
import os

# List of texts to display
texts = ["Text 1", "Text 2", "Text 3", "Text 4", "Text 5"]

# Load images from folder
images = []
for file in os.listdir("/content/images"):
    if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
        image = cv2.imread(os.path.join("/content/images", file))
        images.append(image)

# Define video writer and set codec and frame rate
video = cv2.VideoWriter('video.avi', cv2.VideoWriter_fourcc(*'XVID'), 20.0, (images[0].shape[1], images[0].shape[0]))

# Embed text on each image and write to video
for i, text in enumerate(texts):
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 2
    font_thickness = 2
    text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
    x = (images[i].shape[1] - text_size[0]) // 2
    y = (images[i].shape[0] + text_size[1]) // 2
    bottom_left_corner = (x, y)
    color = (255, 255, 255)
    cv2.putText(images[i], text, bottom_left_corner, font, font_scale, color, font_thickness)
    for j in range(100):
        video.write(images[i])

# Release video writer
video.release()
```

```python
import cv2
import os

# Get list of all images in the folder
images_path = '/content/images/'
images = [os.path.join(images_path, i) for i in os.listdir(images_path) if i.endswith('.jpg')]

# List of texts to be displayed on the images
texts = ["Text 1", "Text 2", "Text 3", "Text 4", "Text 5"]

# Create a VideoWriter object to save the video
fourcc = cv2.VideoWriter_fourcc(*'XVID')
video = cv2.VideoWriter('vidddeo.avi', fourcc, 20.0, (1920, 1080))

# Iterate through each image and add text to it
for i, img_path in enumerate(images):
    img = cv2.imread(img_path)
    height, width, _ = img.shape
    text = texts[i]

    # Add text to the image
    font = cv2.FONT_HERSHEY_SIMPLEX
    text_scale = 1
    text_thickness = 2
    text_size, _ = cv2.getTextSize(text, font, text_scale, text_thickness)
    text_x = int((width - text_size[0]) / 2)
    text_y = int((height + text_size[1]) / 2)
    cv2.putText(img, text, (text_x, text_y), font, text_scale, (255, 255, 255), text_thickness)

    # Write the image to the video
    video.write(img)

# Release the video writer
video.release()
```
