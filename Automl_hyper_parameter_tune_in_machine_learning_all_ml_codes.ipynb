```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# Load the Iris dataset
iris = datasets.load_iris()

# Create a DataFrame from the Iris data
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])

# Assign the target variable as "Species"
df['Species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Split the data into training and testing sets
X = df.iloc[:, :-2].values
y = df.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Fit KNN and Decision Trees models to the data
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

# Predict using KNN and Decision Trees models
y_pred_knn = knn.predict(X_test)
y_pred_dt = dt.predict(X_test)

# Evaluate the models using accuracy score
acc_knn = accuracy_score(y_test, y_pred_knn)
acc_dt = accuracy_score(y_test, y_pred_dt)

# Plot confusion matrix for KNN and Decision Trees models
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_dt = confusion_matrix(y_test, y_pred_dt)

# Print accuracy score and confusion matrix for KNN and Decision Trees models
print("KNN Accuracy:", acc_knn)
print("KNN Confusion Matrix:", cm_knn)
print("Decision Trees Accuracy:", acc_dt)
print("Decision Trees Confusion Matrix:", cm_dt)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import precision_recall_fscore_support, accuracy_score
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Load Iris data
data = pd.read_csv('/content/Iris.csv')

# Split into features and target
X = data.drop(['Species'], axis=1)
y = data['Species']

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Define a base MLP Classifier
mlp = MLPClassifier(max_iter=1000)

# Define a hyperparameter grid to search over
param_grid = {
    'hidden_layer_sizes': [(10,), (50,), (100,)],
    'activation': ['relu', 'tanh', 'logistic'],
    'solver': ['adam', 'lbfgs'],
    'alpha': [0.0001, 0.001, 0.01],
}

# Use RandomizedSearchCV to tune hyperparameters
random_search = RandomizedSearchCV(mlp, param_grid, cv=5, scoring='accuracy', n_iter=20)
random_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best Parameters for MLP Classifier (Random Search): ", random_search.best_params_)
print("Best Score for MLP Classifier (Random Search): ", random_search.best_score_)

# Predict on test set using best parameters
y_pred = random_search.predict(X_test)

# Calculate precision, recall, f1-score, and accuracy
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')
accuracy = accuracy_score(y_test, y_pred)

# Print metrics
print("Precision: ", precision)
print("Recall: ", recall)
print("f1-score: ", f1_score)
print("Accuracy: ", accuracy)
```

```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Load iris dataset
df = pd.read_csv("/content/Iris.csv")

# Define target variable
y = df['Species']

# Drop target variable and set input variables
X = df.drop(['Species'], axis=1)

# Initialize MLPClassifier
mlp = MLPClassifier(max_iter=1000)

# Define hyperparameters to tune
param_grid = {'hidden_layer_sizes': [(10,), (50,), (100,)],
              'alpha': [0.0001, 0.001, 0.01]}

# Perform grid search
grid_search = GridSearchCV(mlp, param_grid, cv=3, scoring='accuracy', return_train_score=True)
grid_search.fit(X, y)

# Print best parameters and best score
print("Best Parameters for MLP (Grid Search): ", grid_search.best_params_)
print("Best Score for MLP (Grid Search): ", grid_search.best_score_)

# Predict using best parameters
mlp = MLPClassifier(max_iter=1000, **grid_search.best_params_)
mlp.fit(X, y)
y_pred = mlp.predict(X)

# Print evaluation metrics
print("Accuracy:", accuracy_score(y, y_pred))
print("Classification Report:")
print(classification_report(y, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y, y_pred))
```

```python
# Importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load data
df = pd.read_csv("/content/Iris.csv")
X = df.iloc[:, :-1]
y = df["Species"]

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create an SVM classifier
svm_clf = svm.SVC()

# Define hyperparameters and possible values for grid search
grid_params = {'C': [0.1, 1, 10, 100], 'kernel': ['linear', 'rbf']}

# Use grid search to tune the hyperparameters
grid_search = GridSearchCV(svm_clf, grid_params, scoring='accuracy', cv=5)
grid_search.fit(X_train, y_train)

# Use randomized search to tune the hyperparameters
random_search = RandomizedSearchCV(svm_clf, grid_params, n_iter=50, scoring='accuracy', cv=5, random_state=0)
random_search.fit(X_train, y_train)

# Make predictions using the best hyperparameters
y_pred_grid = grid_search.predict(X_test)
y_pred_random = random_search.predict(X_test)

# Evaluate the performance using accuracy, precision, recall, and f1 score
grid_acc = accuracy_score(y_test, y_pred_grid)
grid_precision = precision_score(y_test, y_pred_grid, average='weighted')
grid_recall = recall_score(y_test, y_pred_grid, average='weighted')
grid_f1 = f1_score(y_test, y_pred_grid, average='weighted')

random_acc = accuracy_score(y_test, y_pred_random)
random_precision = precision_score(y_test, y_pred_random, average='weighted')
random_recall = recall_score(y_test, y_pred_random, average='weighted')
random_f1 = f1_score(y_test, y_pred_random, average='weighted')

# Print the results
print("Grid Search Results:")
print("Accuracy: ", grid_acc)
print("Precision: ", grid_precision)
print("Recall: ", grid_recall)
print("F1 Score: ", grid_f1)

print("Random Search Results:")
print("Accuracy: ", random_acc)
print("Precision: ", random_precision)
print("Recall: ", random_recall)
print("F1 Score: ", random_f1)
```
